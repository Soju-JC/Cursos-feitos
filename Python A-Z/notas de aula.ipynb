# Notas de aula do curso Python A-Z


## Símbolos

- NaN: Dado faltante
- None: Mesmo que o NULL do R

## Funções

- type(x): Retorna a classe do objeto x (similar ao class() do R)
- int(x): Converte x em int
- str(x): Converte x em string
- float(x): Converte x em float
- len(x): Retorna a quantidade de caracteres caso x for string (seu tamanho), ou tamanho da lista caso x for lista
- input(): Entrada de input do usuário (similar ao readline() do R). Como a entrada é do teclado, sempre será string
- range(x): Sequencia dos x inteiros iniciando do zero
- list(range(x)): Retorna range como uma lista
- randn(): Função do numpy.random que gera um valor pseudo-aleatório normal 

## Booleanos

- and: Equivale ao && do R
- or: Equivalente ao | do R
- not: Equivale ao ! do R
- True: Equivale ao TRUE do R
- False: Equivale ao FALSE do R

## Posicionamento

- x.find("string") ou x.index("string"): Retorna a posição em x da string procurada
- x[position]: Retorna o caracter da string x na posição desejada

## Utilidades

- a // b: Pega o inteiro da divisão de 'a' por 'b' (similar ao %/% do R)
- a % b: Pega o resto da divisão de a por b (similar ao %% do R)
- +: O 'mais' serve tanto para somar números quanto para concatenar textos
- **: Potência (similar ao ^ do R)
- != ou <>: Significa 'diferente de'

##### Library math

- math.sqrt(x): Calcula a raíz quadrada de x
- math.log(x): Calcula o ln de x
- math.exp(x): Calcula exponencial de x (exp(x) ou e(x))
- math.floor(x): Calcula o menor inteiro de x (arredonda pra baixo)
- math.factorial(x): Calcula o fatorial de x
- math.prod(range(1, 5)): Calcula o produtório da sequência de inteiros de 1 a 5 (não inclui o 5)
- math.gamma(x): Calcula a função gamma de x

## Condicionais

#### Nested conditionals

```python
import numpy as np
from numpy.random import randn

n = randn()
if n > 1:
    resposta = 'Maior que 1'
else:
    if n < -1:
        resposta = 'Menor que 1'
    else:
        resposta = 'Ente -1 e 1'
print('n = ', n, ';', 'resposta: ', resposta)
```

#### Chained conditionals

```python
if n > 1:
    resposta = 'Maior que 1'
elif n < -1: 
    resposta = 'Menor que 1'
else: 
    resposta = 'Ente -1 e 1'
print('n = ', n, ';', 'resposta: ', resposta)
```

# Homework 1

Considerando uma distribuição normal padrão (média zero e variância 1), crie um código que verifique a porcentagem de números aleatórios gerados a partir dessa distribuição que caem entre -1 e 1, aumentando o tamanho da amostra. No fim, é esperado algo próximo de 68.2%

```python
# Biblioteca para computação científica.
import numpy as np 
# Gerador de números aleatórios da distribuição normal padrão.
from numpy.random import randn 
# Biblioteca para construir gráficos.
import matplotlib.pyplot as plt 

# Cria lista de tamanhos de amostras (5 a 5000 de 5 em 5).
x = list(range(5, 5001, 5))
# Cria lista nula para guardar as porcentagens.
y = []
# Define uma semente para garantir reproducibilidade.
np.random.seed(5)
# Loop para calcular a porcentagem de valores entre -1 e 1 de cada amostra gerada.
for i in range(0, len(x), 1):
    aux1 =randn(x[i])
    aux2 = aux1[(aux1 > -1) & (aux1 < 1)]
    y.append((len(aux2)/len(aux1)) * 100)

# Chama o método subplots() para gerar o gráfico.    
fig, ax = plt.subplots()
# Cria o gráfico cruzando x e y.
ax.plot(x, y)
# Adiciona segmentos de linhas vermelhos horizontais em y = 68.2 (valor esperado).
ax.axhline(y = 68.2, linestyle = "--", color = "red")
# Define Título e labels.
ax.set_title("Lei dos grandes números")
ax.set_xlabel("\nTamanho da amostra")
ax.set_ylabel("Porcentagem do -1 até 1\n")

# Mostra o gráfico.
plt.show()
```

# Listas

As listas em Python possuem um comportamento similar aos vetores do R. Enquanto no R tudo é vetor, de certa forma, no Python tudo é lista e pode ser manipulado como tal.

## Exemplo

```python
# Lista sem nada dentro
lista = []
# Lista com diferentes objetos dentro
lista = [5, 5.5, 'texto', True]
# Listas podem conter outras listas
lista2 = [5, 5.5, 'texto', True, lista]
# Seleciona o primeiro elemento da lista
lista2[0]
# Seleciona o último elemento da lista
lista2[len(lista2)-1] # Equivale a lista2[4] 
# Seleciona os 4 primeiros elementos da lista
lista2[0:4] # Seleciona todos os elementos entre as posições 0 e 4, mas sem incluir o da posição 4 (ou seja, 0, 1, 2 e 3) 

lista3 = list(range(13))
# Seleciona todos elementos da lista iniciando da posição 0 e não dando pulos entre os elementos
print(lista3[0:len(lista3):1]) # Equivale a lista3[ : ]
# Seleciona todos elementos a partir da posição 5
print(lista3[5:])
# Seleciona todos os 5 primeiros elementos
print(lista3[:5])
# Seleciona todos elementos da lista iniciando da posição 0 e pulando 1 passo
print(lista3[0:len(lista3):2])
# Seleciona todos elementos da lista pulando 1 passo
print(lista3[::2])
```

Os elementos da lista podem ser indexados de trás para frente, onde, o último elemento possui index -1, o penúltimo -2, e assim por diante.

```python
# Seleciona o primeiro elemento da lista
print(lista2[-len(lista2)]) # Equivale a lista2[-5] 
# Seleciona o último elemento da lista
print(lista2[-1]) # Equivale a lista2[4] ou  lista2[len(lista2)-1]
# Seleciona os 4 primeiros elementos da lista
print(lista2[-5:-1]) # Seleciona do primeiro elemento ao quarto 
print(lista2[-5:4]) # Seleciona do primeiro elemento ao quarto (misturei index positivos e negativos)
```

Eventualmente, podemos utilizar esses métodos de seleção de elemento para aplicar alterações à lista.

```python
# Print da lista original
print('Lista original:', lista2)
# Altera o primeiro elemento da lista
lista2[0] = 99
print(lista2)
# Altera o último elemento da lista
lista2[len(lista2)-1] = 'e'
print(lista2)
# Altera os 4 primeiros elementos da lista
lista2[0:4] = ['a', 'b', 'c', 'd'] 
print(lista2)  
```
